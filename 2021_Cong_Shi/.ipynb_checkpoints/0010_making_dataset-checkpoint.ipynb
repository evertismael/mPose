```python
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
import scipy.io
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

device = 'cuda' if torch.cuda.is_available() else 'cpu'
plt.plot([0,1,0,1])
plt.show()
```

```python

class RnAngDataset(Dataset):
    def __init__(self, path_images, batch_id, batch_size):
        self.path_images = path_images
        self.batch_id = batch_id
        self.batch_size = batch_size
        
        m_start = 1 + (self.batch_id-1)*self.batch_size
        m_end = self.batch_id*self.batch_size

        f_name = '/batch_' + str(m_start) + '_' + str(m_end) + '.mat';
        mat = scipy.io.loadmat(self.path_images + f_name)

        self.joints_xyz = mat['jnts_xyz_all']
        self.rm_azm = mat['rm_azm_all']
        self.rm_elm = mat['rm_azm_all']



    def __getitem__(self, idx):
        rm_azm_i = self.rm_azm[:,:,idx]
        rm_elm_i = self.rm_elm[:,:,idx]

        x = np.concatenate((np.expand_dims(rm_azm_i, axis=2),np.expand_dims(rm_azm_i, axis=2)), axis=2)
        y = self.joints_xyz[:,:,idx]
        return x, y
        
    def __len__(self):
        return self.rm_azm.shape[2]

path_images = '../mPoseDataCreator/r_ang_imgs/'
batch_id = 1
batch_size = 20
train_dataset = RnAngDataset(path_images, batch_id, batch_size)
train_loader = DataLoader(train_dataset, batch_size=50, shuffle=True)
```

```python
class mPosNet(nn.Module):
    def __init__(self):
        super().__init__()
        
        # which layers to use
        self.conv2d_1 = nn.Conv2d(in_channels=2, out_channels=16, kernel_size=5, stride=2)
        self.conv2d_2 = nn.Conv2d(in_channels=16, out_channels=32, kernel_size=5, stride=2)
        self.conv2d_3 = nn.Conv2d(in_channels=32, out_channels=64, kernel_size=5, stride=2)
        
        self.fc_1 = nn.Linear(7*7*64,7*7*64)
        self.fc_2 = nn.Linear(7*7*64,7*7*64)

        self.rl_1 = nn.Linear(7*7*64,3*31)
        
        
    def forward(self, x):
        # how the layers are connected
        x1 = F.relu(self.conv2d_1(x))
        x2 = F.relu(self.conv2d_2(x1))
        x3 = F.relu(self.conv2d_3(x2))
        
        x4 = torch.flatten(x3,1)
        
        x5 = F.relu(self.fc_1(x4))
        x6 = F.relu(self.fc_2(x5))

        x7 = self.rl_1(x6)
                    
        return x7
```

```python
# define training function:
def training_epoch(train_loader, model, device):
    optimizer = optim.SGD(model.parameters(),lr=1e-1)
    model = model.to(device)
    train_loss = []
    
    delta_huber = 0.1
    for x_batch, y_batch in train_loader:
        
        x_batch, y_batch = x_batch.type(torch.FloatTensor), y_batch.type(torch.FloatTensor)
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)
        x_batch = x_batch.permute((0,3,1,2))

        model.train()
        y_hat = model(x_batch)

        # ------------------------------------------------------------------------
        # huber loss:
        y_hat = y_hat.view(-1,3,31)
        p_diff = y_batch - y_hat

        #loss = (p_diff**2).mean()
        
        #p_diff = p_diff.type(torch.DoubleTensor)
        p_diff_mag = torch.sqrt(torch.sum(p_diff**2,1))
        loss = p_diff_mag.mean()
        
        # -------------------------------------------------------------------------
        

        loss.backward()

        optimizer.step()
        optimizer.zero_grad()
        #print(loss.item())
        train_loss.append(loss.item())
        
        # Print batch loss
        # print('\t partial train loss (single batch): %f' % (loss.data))
    #print(train_loss)
    return np.mean(train_loss)
```

```python
n_epoch = 10
model = mPosNet().to(device)
loss_epoch = []
```

```python
for epoch_idx in range(n_epoch):
    # train:
    a = training_epoch(train_loader, model, device)
    loss_epoch.append(a)
    print(a)
    
print(type(loss_epoch))
```

```python
import matplotlib.pyplot as plt
plt.plot(list(loss_epoch))
plt.ylabel('loss')
plt.show()
```
